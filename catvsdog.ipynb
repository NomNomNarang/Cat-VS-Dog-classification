```python
train_path = "/content/drive/MyDrive/CatsVsDog/train"
test_path  = "/content/drive/MyDrive/CatsVsDog/test"
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Activation, BatchNormalization
```

```python
##generators : divides images to batch
train_ds=keras.utils.image_dataset_from_directory (
    directory='/content/drive/MyDrive/CatsVsDog/train_split',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
test_ds=keras.utils.image_dataset_from_directory(
    directory='/content/drive/MyDrive/CatsVsDog/test_split',
    labels='inferred',
    label_mode='int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
#normalize
def process(image,label):
  image=tf.cast(image/255. ,tf.float32)
  return image,label

train_ds=train_ds.map(process)
test_ds=test_ds.map(process)
```

```python
##CNN model
model= keras.Sequential([
    Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu', input_shape=(256,256,3)),
    MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'),
    Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'),
    MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'),
    Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'),
    MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(64,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(train_ds,epochs=10,validation_data=test_ds)
```

```python
y_predict=model.predict(test_ds)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
import numpy as np
y_pred=np.argmax(y_predict,axis=1)
```
